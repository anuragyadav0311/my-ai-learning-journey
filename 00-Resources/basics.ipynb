### Creating Virtual Environment

1. When we first start we should always make venv(virtual environment) using <br>
`conda create -p venv python==3.12`    <br>#or which every python version you desire <br>#in place of venv we can write any name we want to give to the virtual environment
2. Secondly activate the venv by<br>
`conda activate address_of_venv` (address of venv if you are inside project folder is `./venv`)  <br>`conda deactivate` (to deactivate venv) <br>
NOTE: Since i use omarchy there the hashing of the terminal will not allow venv to activate so we need to run 
`set -h` 
before activating venv
3. Then you can also have requirement.txt where you will mention desire external libraries and packages
`pip install -r requirement.txt`

### Another way for venv (without conda):

1. Navigate to Your Project Directory
Open your terminal and cd into the folder where your project lives:

`cd /path/to/your/project`

2. Create the Virtual Environment
Run this command to generate the environment. The second venv is the name of the folder that will be created (this is the industry standard naming convention):

`python3 -m venv venv`

3. Activate the Environment
Before you install any packages or run your code, you must activate the environment so your terminal knows to use it instead of the global system Python:

`source venv/bin/activate`  # for linux/mac <br>
`venv\Scripts\activate`     # for windows

(You will know it is working because your terminal prompt will change to show (venv) at the beginning of the line).
4. Install Your Packages
Now you can safely install dependencies using pip. They will be isolated entirely to this project folder:

`pip install package_name`

5. Deactivate (When Finished)
Once you are done working on the project and want to return to your normal system environment, simply run:

`deactivate`
